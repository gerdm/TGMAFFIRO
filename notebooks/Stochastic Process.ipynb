```python
import numpy as np
from scipy.stats import norm
from numpy.random import seed
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

%matplotlib notebook
```

```python
seed(1643)

along = np.ones(150)
fig = plt.figure(figsize=(8,5))
ax = fig.add_subplot(111, projection='3d')

time = np.arange(0.1, 1.1, 0.1)
for t in time:
    xnorm = np.linspace(-3, 3, 150)
    ynorm = norm.pdf(xnorm, scale = np.sqrt(t))
    ax.plot(t * along, xnorm, ynorm, color="teal", linewidth=1.5, alpha=0.9)

T = np.linspace(0.1, 1, 150)
Xt = np.random.normal(scale=np.sqrt(T))
ax.plot(T, Xt, linewidth=1, color="crimson")

plt.gca().patch.set_facecolor('white')
ax.w_xaxis.set_pane_color((0, 0, 0, 0))
ax.w_yaxis.set_pane_color((0, 0, 0, 0))
ax.w_zaxis.set_pane_color((0, 0, 0, 0))

ax.set_ylim(-3.5, 3.5)
ax.view_init(32, -133)

ax.set_xlabel("$t$")
ax.set_ylabel("$X_t$")
ax.set_zlabel("$P(X_t)$")

plt.savefig("../images/stoch_process_3d.pdf")
```
