# Lebesgue Integrals 

```python
import numpy as np
from numpy import vectorize as v
import matplotlib.pyplot as plt
```

```python
f = v(lambda x: x ** 2)
xran = np.linspace(-2, 2, 1000)
A = 1 * (f(xran) >= 1)

plt.plot(xran, f(xran), label="$f(x)$")
plt.scatter(xran, A, s=1, c="orange", label="$1_A$")
plt.show()
```
